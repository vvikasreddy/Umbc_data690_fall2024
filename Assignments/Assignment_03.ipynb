Let's experiment with column "10. What do you expect to get out of this course?" to see if we can determine most popular leisure time actvities.

````python
list_answers = df["9. What do you enjoy doing in your leisure time?"].tolist()

list_answers
````

We will start with an empty dictionary like this:
```python
word_freq = {}
```

We hope to end up with something like this:
```python
word_freq{
    "movies" : 3,
    "reading": 10,
    "sports": 3
}
```

````python
word_freq = {}

for answer in list_answers:
    list_words = answer.split()
    for word in list_words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

word_freq
````

## Assignment 03 Part 2

Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

# Part 2 answer

````python
career_goals = df["11. What is your long-term career goal?"].tolist()

career_goals
````

````python
# trying to find the frequencies of words

words_freq = {}

for goal in career_goals:
    list_words = goal.split()
    for word in list_words:
        if word in words_freq:
            words_freq[word] += 1
        else:
            words_freq[word] = 1
````

````python
# sorting the words by frequency

sorted(words_freq.items(), key=lambda x: x[1], reverse=True)
````

# Insights from the analysis:

- The first meaningfull word that appears in the sentences, is the word "data", which is not surprising given this is a course in data science. We can infer that most people are dreaming of data related careers.
- When parsed by individual words, we can see that most of the words appearing seem to be the stopping words i.e (the, a, an, am etc.,)
- If we neglect all those stop words, the next common words give more information, they are analyst, machine, technology, advanced, analytics, machine, decision-making, organization, projects, solve and leadership, these words convey a lot of information.
- To sum up, most of the people are craving for technological, leadership and data related roles.
